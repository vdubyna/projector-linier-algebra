### 2. Знайти найближчий вектор для кожного вектора з групи векторів: 
Відстані між векторами вимірюються за трьома різними метриками: $L_1, L_2, L_\infty$.

* $\vec{a} = (0, -1, 1)$
* $\vec{b} = (0, 4, 0)$
* $\vec{c} = (2, 1, 4)$ 

Для знаходження найменших відстаней використаєм наступні формули:

1.	$L_1$ - норма (мангеттенська відстань):  
$d_{L_1}(\vec{x}, \vec{y}) = \sum_{i} |x_i - y_i|$  
Тут ми знаходимо спочатку вектор між векторами x та y,  
а потім додаєм всі елементи по модулю.
2.	$L_2$ - норма (евклідова відстань):  
$d_{L_2}(\vec{x}, \vec{y}) = \sqrt{\sum_{i} (x_i - y_i)^2}$  
Тут ми вираховуєм відстань використовуючи теорему Піфігора.
3. $L_\infty$ - норма (максимальна відстань):  
$d_{L_\infty}(\vec{x}, \vec{y}) = \max_i |x_i - y_i|$  
Тут ми знаходимо спочатку вектор між векторами x та y,
а потім берем найбільший елемент вектора по модулю 

Відстані між векторами a та b:  
1. $L1(a, b) = |0 - 0| + |-1 - 4| + |1 - 0| = 0 + 5 + 1 = 6$
2. $L2(a, b) = \sqrt{(0 - 0)^2 + (-1 - 4)^2 + (1 - 0)^2} = \sqrt{0 + 25 + 1} = \sqrt{26} \approx 5.10$
3. $L\infty(a, b) = \max(|0 - 0|, |-1 - 4|, |1 - 0|) = \max(0, 5, 1) = 5$

Відстані між векторами a та c:
1. $L1(a, c) = |0 - 2| + |-1 - 1| + |1 - 4| = 2 + 2 + 3 = 7$
2. $L2(a, c) = \sqrt{(0 - 2)^2 + (-1 - 1)^2 + (1 - 4)^2} = \sqrt{4 + 4 + 9} = \sqrt{17} \approx 4.12$
3. $L\infty(a, c) = \max(|0 - 2|, |-1 - 1|, |1 - 4|) = \max(2, 2, 3) = 3$

Відстані між векторами  b  та  c:
1. $L1(b, c) = |0 - 2| + |4 - 1| + |0 - 4| = 2 + 3 + 4 = 9$
2. $L2(b, c) = \sqrt{(0 - 2)^2 + (4 - 1)^2 + (0 - 4)^2} = \sqrt{4 + 9 + 16} = \sqrt{29} \approx 5.39$
3. $L\infty(b, c) = \max(|0 - 2|, |4 - 1|, |0 - 4|) = \max(2, 3, 4) = 4$

Згідно з отриманими результатами:
* За L1-метрикою:
  * Для  a  найближчий вектор —  b  (відстань 6).
  * Для  b  найближчий вектор —  a  (відстань 6).
  * Для  c  найближчий вектор —  a  (відстань 7).
* За L2-метрикою:
  *	Для  a  найближчий вектор —  c  (відстань приблизно 4.12).
  *	Для  b  найближчий вектор —  a  (відстань приблизно 5.10).
  *	Для  c  найближчий вектор —  a  (відстань приблизно 4.12).
* За L∞-метрикою:
  *	Для  a  найближчий вектор —  c  (відстань 3).
  *	Для  b  найближчий вектор —  c  (відстань 4).
  *	Для  c  найближчий вектор —  a  (відстань 3).

### 3. For x from x1=0 to x2=8 plot the function y using KNN. 
Take k=2 and the dataset of pairs (x, y).  
data = [ (0, 1), (3, 5), (4, 3), (7, 4) ]

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor

# Given dataset of pairs (x, y)
data = np.array([(0, 1), (3, 5), (4, 3), (7, 4)])
# Виділення x-значень (перший елемент у кожній парі) і зміна форми для сумісності з KNN
x_data = np.array([pair[0] for pair in data]).reshape(-1, 1)
# Виділення y-значень (другий елемент у кожній парі)
y_data = np.array([pair[1] for pair in data])

# Set parameters for KNN
k = 2
knn = KNeighborsRegressor(n_neighbors=k)
knn.fit(x_data, y_data)

# Generate x values from 0 to 8 for prediction
x_values = np.linspace(0, 8, 100).reshape(-1, 1)
y_pred = knn.predict(x_values)

# Plotting the function y using KNN
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_pred, label="KNN Prediction (k=2)", color="blue")
plt.scatter(x_data, y_data, color="red", label="Data Points")
plt.xlabel("x")
plt.ylabel("y")
plt.title("KNN Prediction for y using k=2")
plt.legend()
plt.grid(True)
plt.show()
```

### 5. Optional: Find clusters in the dataset contained in clusters.csv file. You can use k-means or any other available clustering methods.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Завантаження даних із файлу
data = pd.read_csv('lesson/data/1713266377370_clusters.csv', header=None)
data.columns = ['Feature1', 'Feature2', 'Feature3']

# Використовуємо метод "лікоть" для визначення оптимальної кількості кластерів
inertia = []
K = range(2, 10)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    inertia.append(kmeans.inertia_)

# Візуалізація методу "лікоть"
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(K, inertia, 'bo-')
plt.xlabel('Кількість кластерів')
plt.ylabel('Inertia')
plt.title('Метод "лікоть" для вибору кількості кластерів')

# Вибір оптимального значення к (наприклад, 3 кластери) та побудова моделі K-means
optimal_k = 3  # Це значення можна скоригувати, виходячи з графіків
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
data['Cluster'] = kmeans.fit_predict(data)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['Feature1'], data['Feature2'], data['Feature3'], c=data['Cluster'], cmap='viridis')
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
plt.title(f'Кластеризація з k = {optimal_k}')
plt.show()
```

### 6. Optional: Encyclopedia of Chess Openings includes about one thousand codes.

In the dataset in chess_openings.csv, each opening is represented by a
vector. Vectors are in the rows of the dataset. For your convenience there is
another csv file (debuts.csv) that contains the names of the openings and
their codes. For the opening A54 - "Old Indian, Ukrainian Variation, 4.Nf3
(A54)”, find 10 nearest vectors (similar openings) for both L1 and L2 distance
metrics.

```python
import pandas as pd
import numpy as np
from scipy.spatial import distance

# Завантаження даних з файлів
chess_vectors = pd.read_csv('lesson/data/1713266383888_chess openings.csv', header=None)
debuts = pd.read_csv('lesson/data/1713266390567_debuts.csv', header=None, names=['Code', 'Name'])

# Знаходження вектора для A54
a54_row = debuts[debuts['Code'] == 'A54'].index[0]
a54_vector = chess_vectors.iloc[a54_row, 1:].values.astype(float)

# Обчислення L1 та L2 відстаней між A54 і всіма іншими векторами
distances_l1 = []
distances_l2 = []

for i, row in chess_vectors.iterrows():
    vector = row[1:].values.astype(float)

    # L1 distance
    l1_dist = np.sum(np.abs(a54_vector - vector))
    distances_l1.append((i, l1_dist))

    # L2 distance
    l2_dist = distance.euclidean(a54_vector, vector)
    distances_l2.append((i, l2_dist))

# Сортуємо список відстаней за відстанню (другий елемент кортежу)
sorted_distances_l1 = sorted(distances_l1, key=lambda x: x[1])
sorted_distances_l2 = sorted(distances_l2, key=lambda x: x[1])

# Виключаємо сам вектор A54 та беремо 10 найближчих
nearest_l1 = sorted_distances_l1[1:11]
nearest_l2 = sorted_distances_l2[1:11]

# Отримання назв дебютів для найближчих векторів
nearest_l1_names = [(debuts.iloc[i]['Code'], debuts.iloc[i]['Name'], dist) for i, dist in nearest_l1]
nearest_l2_names = [(debuts.iloc[i]['Code'], debuts.iloc[i]['Name'], dist) for i, dist in nearest_l2]

# Виведення результатів
print("10 найближчих дебютів за метрикою L1:")
for code, name, dist in nearest_l1_names:
    print(f"{code} - {name}, Відстань L1: {dist}")

print("\n10 найближчих дебютів за метрикою L2:")
for code, name, dist in nearest_l2_names:
    print(f"{code} - {name}, Відстань L2: {dist}")
```
